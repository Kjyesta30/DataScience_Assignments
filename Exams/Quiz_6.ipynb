```python
# Importing Libraries
import numpy as np
import os
import pandas as pd
import tensorflow as tf
```

```python
#Loading the data
housing = pd.read_csv("housing.csv")
housing.head()
```

```python
#Dropping the missing values
housing.dropna(axis=0, inplace=True)
```

```python
# Target Variables and Input variables
housing_X = housing.drop("ocean_proximity", axis=1) # drop labels 
housing_target = housing["ocean_proximity"]
housing_target.value_counts()
```

```python
#Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
housing_X_std = scaler.fit_transform(housing_X)
```

```python
# Spliiting Train,Test
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(housing_X_std, housing_target, test_size=0.5)
```

**MLP Classfier without early stopping**

```python
from sklearn.neural_network import MLPClassifier
mlp_clf = MLPClassifier(hidden_layer_sizes=(15,10,7),max_iter=500,verbose=True)
mlp_clf.fit(train_x, train_y)
```

```python
mlp_clf.n_iter_
```

We can see that the iterating has stopped at 288 as the training loss did not improve after the iteration at tolerance=0.000100 for 10 epochs.

```python
test_y_pred = mlp_clf.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, test_y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_y, test_y_pred))
```

The accuarcy for the MLP classifier without early stopping is 90.4%.

**MLP with Early Stopping**

```python
mlp_clf2 = MLPClassifier(hidden_layer_sizes=(15,10,7),max_iter=500,verbose=True,early_stopping=True,)
mlp_clf2.fit(train_x, train_y)
```

```python
mlp_clf2.n_iter_
```

```python
test_y_pred = mlp_clf2.predict(test_x)
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, test_y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_y, test_y_pred))
```

We can see that the iterating has stopped at 71 and the accuracy is 88.0%

From the above, we can say that the early stopping did not help here as the accuracy did not increase.

**KERAS MODEL**

```python
housing2 = pd.read_csv("housing.csv")
#Drop the missing values
housing2.dropna(axis=0, inplace=True)
```

```python
#Drop the missing values
housing2.dropna(axis=0, inplace=True)
housing2.reset_index(inplace=True, drop=True)
housing2.describe()
```

```python
housing_num = housing2.drop("ocean_proximity", axis=1)
housing_target = housing2[["ocean_proximity"]]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
housing_num_std = scaler.fit_transform(housing_num)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
housing_labels_ord = ordinal_encoder.fit_transform(housing_target)
housing_labels_ord[:10]
```

```python
housing_labels_int = housing_labels_ord.astype(int)
housing_labels_int.dtype
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(housing_num_std, housing_labels_int, test_size=0.5)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(15, input_dim=9, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(7, activation='relu'))
model.add(Dense(5, activation='softmax'))
sgd = keras.optimizers.SGD(lr=0.05)
model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
model.fit(train_x, train_y, epochs=100, batch_size=100)
```

```python
# evaluate the model
scores = model.evaluate(test_x, test_y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

We can say that the MLP model and keras model performed almost same as the accuracy is around 90.4 % and the accuracy for the keras model is 90.2%.
